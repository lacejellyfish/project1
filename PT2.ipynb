### Rename columns starting with: "What motivated you to seek a computing class at CCM? "
* Define a function that cleans column names and takes in a column and prefix
* Define the prefix to strip
* Pass the function to all column names to rename the columns with the selected prefix
* Remove apostrophes from any columns

```python
#Define the function
def clean_column(col, prefix):
    if col.startswith(prefix) and '[' in col and ']' in col:
        bracket_content = col.split('[')[-1].split(']')[0] #Selects content inside the brackets
        cleaned = bracket_content.lower().replace(" ", "_").replace("/", "_") #Forces lowercase, replaces spaces and slashes with underscores
        return cleaned #Return the new column (lowercase, underscores, no spaces or slashes, removing irrelevant words)
    return col #If the column is already "cleaned", return it as is

#Define the prefix to strip
prefix = "What motivated you to seek a computing class at CCM? "

#Pass the function to all column names (RENAMES THESE COLUMNS)
new_survey.columns = [clean_column(col, prefix) for col in new_survey.columns]

#Remove apostrophe from column
new_survey.rename(columns={"it’s_a_required_class_for_the_degree_i’m_seeking": "required_class"}, inplace=True)
```

### Clean column: 'current_major'
* Create a map that includes all majors listed
* Create a function that optimizes text values for pandas
* Apply the function to the 'current_major' column
* Display the updated column

```python
#Defines a map that sorts similar majors into umbrella categories
major_map = {
    "child_and_family_studies": "human_services",
    "early_childhood_education": "human_services",
    "communication": "communications",
    "journalism": "communications",
    "criminal_justice": "criminal_justice",
    "exercise_science": "health_sciences",
    "public_health": "health_sciences",
    "nursing": "health_sciences",
    "radiography": "health_sciences",
    "public_administration": "public_services",
    "liberal_arts": "liberal_arts",
    "music_technology": "arts_technology",
    "theatre_technology": "arts_technology",
    "not_in_a_degree_program": "undeclared",
}

#Defines a function to clean and map a major
def clean_major(value):
    #'isinstance' was borrowed from Google. This function returns true if the variable (value) matches the object type string).
    if not isinstance(value, str):
        return "unknown" #If the text is not a string, return 'unknown'
    clean_val = value.strip().lower().replace(" ", "_") #Removes whitespace, forces lowercase, and replaces spaces with lowercase
    return major_map.get(clean_val, "other")

# Use map to clean the column
survey_2["current_major"] = survey_2["current_major"].map(clean_major)
```

### Clean column: 'race'
* Create a map that includes all races listed and condensed
* Create a function that optimizes text values for pandas
* Apply the function to the column
* Display the updated column

```python
#Map dictionary condenses names
race_map = {
    "hispanic_or_latino": "hispanic",
    "american_indian_native_american_alaska_native": "native_american",
    "asian": "asian",
    "black_african_american": "black",
    "native_hawaiian_other_pacific_islander": "pacific_islander",
    "white_caucasian": "white",
    "multi_racial": "multi_racial",  # keep as-is
    "choose_not_to_reply": "no_reply"
}

#Defines function to clean and condense race values
def clean_race(value):
    if pd.isna(value):  #If there are missing values, convert them to no_reply
        return "no_reply"
    items = value.lower().replace(" ", "_").replace("/", "_").split(";") #Optimize text to pandas format and split at semicolon (used for multiple options at once)
    cleaned_items = [race_map.get(item, item) for item in items if item]

    #Remove duplicates
    cleaned_items = list(set(cleaned_items))

    #Some values include more than one race, which evaded mapping attempts.
    #If more than 1 distinct race, mark as multi_racial
    if len(cleaned_items) > 1:
        return "multi_racial"
    elif len(cleaned_items) == 1:
        return cleaned_items[0]
    else:
        return "no_reply"

#Apply map to column
survey_2["race"] = survey_2["race"].map(clean_race)
```

### Clean column: 'gender'
* Create a map that includes all genders listed and condensed
* Create a function that optimizes text values for pandas
* Apply the function to the column
* Display the updated column

```python
#Defines a function that cleans the column values
def clean_gender(value):
    #Map that formats values to match pandas format and condenses refusal options into single category, "no_reply"
    mapping = {
        "Woman": "woman",
        "Man": "man",
        "Non-binary": "non_binary",
        "I do not identify": "no_reply",
        "Prefer not to say": "no_reply"
    }
    return mapping.get(value, value.lower().replace(" ", "_"))  #Return the string with forced formatting incase mapping does not work

#Apply map to the column
survey_2["gender"] = survey_2["gender"].map(clean_gender)
```

### Clean column: 'age'
* Create a map that includes all ages listed and condensed
* Create a function that optimizes text values for pandas
* Apply the function to the column
* Display the updated column

```python
#Map that condenses ages
age_map = {
    "18 and younger": "18_and_below",
    "65+": "65_plus"
}

#Defines a function to clean text values in age column
def clean_age(value):
    #Uses the map defined above
    value = age_map.get(value, value)
    #Forcefully optimizes text to pandas format if map fails
    return value.lower().replace(" ", "_").replace("+", "plus").replace("-", "_")

#Apply the map 
survey_2["age"] = survey_2["age"].map(clean_age)
```
